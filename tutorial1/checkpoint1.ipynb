<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable.

Use this space to modify or create features.

```python
# @@@ 5
#Standardize Total Expenditure per Student
df['TOTAL_EXPENDITURE_PER_STUDENT'] = df['TOTAL_EXPENDITURE'] / df['ENROLL_ALL']
df
```

Feature engineering justification: **With each state in the United States varying in population sizes and student enrollment, I found that standardizing the Total Expenditure for education of a given state based on the amount of students enrolled allows for a fair comparision between states of drastically different enrollment sizes.**

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
# @@@ 8

#The input data for this section will be the Total Expenditure per Student and 4th Grade Reading Score
X = df[["TOTAL_EXPENDITURE_PER_STUDENT", "AVG_READING_4_SCORE"]].dropna()
#8th Grade Math Score is our response variable
y = df.loc[X.index]["AVG_MATH_8_SCORE"]
```

```python
# @@@ 9
#Similarly to Tutorial 1, we are doing a 70/30 split between training data and testing data
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# @@@ 10
from sklearn.linear_model import LinearRegression
```

```python
# @@@ 11
# Create Model
model = LinearRegression()
```

```python
#Input the training data to the model
model.fit(X_train, y_train)
```

```python
#Use the model to make a prediction of y values
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
# @@@ 12
#Calculate R^2 value
print(model.score(X_test, y_test))
#Calculate mean error
print(np.mean(np.abs(model.predict(X_test)-y_test)))
#Calculate mean absolute error
print(np.mean((model.predict(X_test)-y_test)**2)**0.5)
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
# @@@ 13

col_name = 'TOTAL_EXPENDITURE_PER_STUDENT'


f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel('AVG_MATH_8_SCORE')
plt.title("Model Behavior On Training Set")
```
